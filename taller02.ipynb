## Parte Proporcionada

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 400)

y = f(x)

plt.figure(figsize=(8 ,6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
from scipy.optimize import newton

x0 = 100
points = []
x_r = newton(
    f, x0, args=(points, True), maxiter=100
)   # Este no es mtodo de Newton, es secante!
```

Definir la funcin y su derivada para el mtodo de Newton

```python
def evaluate_function(x: float, points: list = None, display_result: bool = False) -> float:
    """
    Evala la funcin f(x) = x^3 - 3x^2 + x - 1 en un punto x dado.
    
    Args:
        x (float): El valor en el que se evala la funcin.
        points (list, optional): Una lista para almacenar los puntos (x, f(x)).
        display_result (bool, optional): Si es True, imprime el resultado de la evaluacin.
    
    Returns:
        float: El resultado de la evaluacin de la funcin en x.
    """
    # Calcular el valor de la funcin
    result = x**3 - 3 * x**2 + x - 1
    
    # Si se proporciona una lista de puntos, agregar el punto (x, f(x))
    if points is not None:
        points.append((x, result))
    
    # Si se solicita, mostrar el resultado en la consola
    if display_result:
        print(f"f({x}) = {result}")
    
    return result

def derivative_of_function(x: float) -> float:
    """
    Calcula la derivada de la funcin f(x) = x^3 - 3x^2 + x - 1 en un punto x dado.
    
    Args:
        x (float): El valor en el que se evala la derivada.
    
    Returns:
        float: El valor de la derivada en x.
    """
    return 3 * x**2 - 6 * x + 1  # Derivada de f(x)
```

Dibujo de la ecuacion

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar un rango de valores x desde -2 hasta 4, con 400 puntos intermedios
x_values = np.linspace(-2, 4, 400)

# Evaluar la funcin f en cada uno de los valores de x
y_values = evaluate_function(x_values)

# Configuracin de la figura para el grfico
plt.figure(figsize=(8, 6))

# Graficar la funcin f(x)
plt.plot(x_values, y_values, label="$f(x) = x^3 - 3x^2 + x - 1$", color='blue')

# Dibujar lneas horizontales y verticales en y=0 y x=0 respectivamente
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

# Agregar una cuadrcula al grfico
plt.grid(color="gray", linestyle="--", linewidth=0.5)

# Ttulos y etiquetas del grfico
plt.title("Grfico de la ecuacin $f(x) = x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")

# Mostrar la leyenda
plt.legend()

# Mostrar el grfico
plt.show()
```

Funcion para probar varias estimaciones iniciales

```python
from scipy.optimize import newton

def find_root_and_plot(initial_guess: float, color: str) -> float:
    """
    Encuentra la raz de la funcin utilizando el mtodo de Newton y grafica la trayectoria de las estimaciones.

    Args:
        initial_guess (float): Valor inicial desde el cual comenzar a buscar la raz.
        color (str): Color para la representacin grfica de la trayectoria.

    Returns:
        float: La raz encontrada, o None si no se encontr convergencia.
    """
    # Lista para almacenar los puntos (x, f(x)) durante la bsqueda de la raz
    points = []
    
    try:
        # Usar el mtodo de Newton con la funcin y su derivada
        root = newton(evaluate_function, initial_guess, fprime=derivative_of_function, 
                      args=(points, False), maxiter=100, full_output=True, disp=False)
        
        # Extraer la raz encontrada
        root_value = root[0]
        print(f"Raz encontrada desde x0 = {initial_guess}: x = {root_value:.4f}")
    
    except RuntimeError:
        # Manejar el caso en que no se encuentra convergencia
        root_value = None
        print(f"No se encontr convergencia desde x0 = {initial_guess}")
    
    # Extraer las estimaciones de los puntos
    x_points = [point[0] for point in points]
    y_points = [point[1] for point in points]

    # Graficar la trayectoria de las estimaciones de la raz
    plt.plot(x_points, y_points, "o-", label=f"Trayectoria desde x0 = {initial_guess}", color=color)
    
    return root_value
```
