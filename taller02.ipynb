## Parte Proporcionada

```python
def f(x: float, points: list[float] = [], show: bool = False):
    y = x**3 - 3 * x**2 + x - 1
    points.append((x, y))
    if show:
        print(f"f({x}) = {y}")
    return y
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 4, 400)

y = f(x)

plt.figure(figsize=(8 ,6))
plt.plot(x, y, label="$x^3 - 3x^2 + x - 1$")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of the equation $x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

```python
from scipy.optimize import newton

x0 = 100
points = []
x_r = newton(
    f, x0, args=(points, True), maxiter=100
)   # Este no es método de Newton, es secante!
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton
```

Definir la función y su derivada para el método de Newton

```python
def evaluate_function(x: float, points: list = None, display_result: bool = False) -> float:
    """
    Evalúa la función f(x) = x^3 - 3x^2 + x - 1 en un punto x dado.
    
    Args:
        x (float): El valor en el que se evalúa la función.
        points (list, optional): Una lista para almacenar los puntos (x, f(x)).
        display_result (bool, optional): Si es True, imprime el resultado de la evaluación.
    
    Returns:
        float: El resultado de la evaluación de la función en x.
    """
    # Calcular el valor de la función
    result = x**3 - 3 * x**2 + x - 1
    
    # Si se proporciona una lista de puntos, agregar el punto (x, f(x))
    if points is not None:
        points.append((x, result))
    
    # Si se solicita, mostrar el resultado en la consola
    if display_result:
        print(f"f({x}) = {result}")
    
    return result

def derivative_of_function(x: float) -> float:
    """
    Calcula la derivada de la función f(x) = x^3 - 3x^2 + x - 1 en un punto x dado.
    
    Args:
        x (float): El valor en el que se evalúa la derivada.
    
    Returns:
        float: El valor de la derivada en x.
    """
    return 3 * x**2 - 6 * x + 1  # Derivada de f(x)
```

Dibujo de la ecuacion

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar un rango de valores x desde -2 hasta 4, con 400 puntos intermedios
x_values = np.linspace(-2, 4, 400)

# Evaluar la función f en cada uno de los valores de x
y_values = evaluate_function(x_values)

# Configuración de la figura para el gráfico
plt.figure(figsize=(8, 6))

# Graficar la función f(x)
plt.plot(x_values, y_values, label="$f(x) = x^3 - 3x^2 + x - 1$", color='blue')

# Dibujar líneas horizontales y verticales en y=0 y x=0 respectivamente
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)

# Agregar una cuadrícula al gráfico
plt.grid(color="gray", linestyle="--", linewidth=0.5)

# Títulos y etiquetas del gráfico
plt.title("Gráfico de la ecuación $f(x) = x^3 - 3x^2 + x - 1$")
plt.xlabel("x")
plt.ylabel("f(x)")

# Mostrar la leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

Funcion para probar varias estimaciones iniciales

```python
from scipy.optimize import newton

def find_root_and_plot(initial_guess: float, color: str) -> float:
    """
    Encuentra la raíz de la función utilizando el método de Newton y grafica la trayectoria de las estimaciones.

    Args:
        initial_guess (float): Valor inicial desde el cual comenzar a buscar la raíz.
        color (str): Color para la representación gráfica de la trayectoria.

    Returns:
        float: La raíz encontrada, o None si no se encontró convergencia.
    """
    # Lista para almacenar los puntos (x, f(x)) durante la búsqueda de la raíz
    points = []
    
    try:
        # Usar el método de Newton con la función y su derivada
        root = newton(evaluate_function, initial_guess, fprime=derivative_of_function, 
                      args=(points, False), maxiter=100, full_output=True, disp=False)
        
        # Extraer la raíz encontrada
        root_value = root[0]
        print(f"Raíz encontrada desde x0 = {initial_guess}: x = {root_value:.4f}")
    
    except RuntimeError:
        # Manejar el caso en que no se encuentra convergencia
        root_value = None
        print(f"No se encontró convergencia desde x0 = {initial_guess}")
    
    # Extraer las estimaciones de los puntos
    x_points = [point[0] for point in points]
    y_points = [point[1] for point in points]

    # Graficar la trayectoria de las estimaciones de la raíz
    plt.plot(x_points, y_points, "o-", label=f"Trayectoria desde x0 = {initial_guess}", color=color)
    
    return root_value
```
